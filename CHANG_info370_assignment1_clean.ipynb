```python
# Find and remove the words.
dataClean['ingredients'] = dataClean['ingredients'].apply(multiple_replace)
```

```python
# Find - and () and remove them.
def replaceIt(df):
    df = re.sub(r",.(\w+).*", "", df)
    df = re.sub(r"Or\s(\w+).*", "", df)
    df = re.sub(r"(\w+)\-(\w+)\s","", df)
    df = re.sub(r"\-(\w+)\s","", df)
    df = re.sub(r"\(.*\)","", df)
    return df
```

```python
# Find and replace characters and spacings so the data is clean
dataClean['ingredients'] = dataClean['ingredients'].apply(replaceIt)
dataClean['ingredients'] = dataClean['ingredients'].str.replace(r"^\s",'')
```
